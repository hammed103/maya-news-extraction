```python
import json
import re

def extract_news_stories(data_string):
    # Find complete story objects
    stories = []
    
    # Look for story objects with title, description, bias data, etc.
    pattern = r'"biasSourceCount":(\d+).*?"title":"([^"]+)".*?"description":"([^"]+)".*?"summary":"([^"]+)"'
    matches = re.findall(pattern, data_string, re.DOTALL)
    
    for match in matches:
        source_count, title, description, summary = match
        
        # Find bias data for this story
        bias_pattern = r'"coverageProfileStatement":"([^"]+)".*?"leftPercent":(\d+),"rightPercent":(\d+),"centerPercent":(\d+)'
        bias_match = re.search(bias_pattern, data_string)
        
        # Find URL for this story
        url_pattern = r'"slug":"([^"]+)"'
        url_match = re.search(url_pattern, data_string)
        url = f"https://ground.news/{url_match.group(1)}" if url_match else "URL not found"
        
        # Find headline (generated headline might be different from title)
        headline_pattern = r'"generatedHeadline":"([^"]+)"'
        headline_match = re.search(headline_pattern, data_string)
        headline = headline_match.group(1) if headline_match else title
        
        bias_info = {}
        if bias_match:
            bias_info = {
                "statement": bias_match.group(1),
                "left": bias_match.group(2),
                "right": bias_match.group(3),
                "center": bias_match.group(4)
            }
        
        # Clean up text (handle special characters)
        title = title.replace("â€™", "'").replace("â€œ", '"').replace("â€", '"')
        description = description.replace("â€™", "'").replace("â€œ", '"').replace("â€", '"')
        summary = summary.replace("â€™", "'").replace("â€œ", '"').replace("â€", '"')
        headline = headline.replace("â€™", "'").replace("â€œ", '"').replace("â€", '"') if headline else ""
        
        stories.append({
            "title": title,
            "headline": headline,
            "description": description,
            "summary": summary,
            "url": url,
            "source_count": source_count,
            "bias_info": bias_info
        })
    
    return stories

# Extract the news stories
news_stories = extract_news_stories(response.text)

# Print the extracted information in a readable format
print("=== ISRAEL-IRAN CONFLICT NEWS SUMMARY ===\n")

for i, story in enumerate(news_stories):
    print(f"STORY {i+1}:")
    print(f"Title: {story['title']}")
    print(f"Headline: {story['headline']}")
    print(f"URL: {story['url']}")
    print(f"Summary: {story['summary']}")
    print(f"Sources: {story['source_count']} news outlets covered this story")
    
    if story['bias_info']:
        print(f"Media Bias: {story['bias_info']['left']}% Left, {story['bias_info']['right']}% Right, {story['bias_info']['center']}% Center")
    
    print("-" * 70)
```
